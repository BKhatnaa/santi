# Neural network

Basic machine takes a question, does some “thinking” and pushes out an answer. Just like human's brain.

Neurons are the fundamental units of the brain. They transmit an electrical signal from one end to the other.Computer has many inputs and outputs just like neurons.That is why we call neural network.

Matrices allow us to express the work we need to do concisely and easily, and computers can get the calculations done quickly and efficiently. Therefore we should convert layers to matrices.

Gradient descent is faster way to find best option.Gradient descent starts randomly chosen starting point and differentiate find out what the slope is in that specific point and find out if slope is positive or negative. Calculate the slope if slope is negative, go right.If slope is positive go left.
