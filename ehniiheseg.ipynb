### Data tampering

Exchange files through unsecured network traffic is very important issue.
The files can be deleted, edited or copied which is called data tampering.
When party A sends file to party B it can be tampered 3 cases:
    1. Interception
        There is possibility that message contents are intercepteed by party X during transmission. It can be secured by **Encryption** technology.
    2. Spoofing
        There's a chance that party X was posing as party B.
        It can be secured by **Message authentication code**
    3. Falsification
        There's a chance that party X had overwritten the message during transmission.
        It can be secured by **Digital signature**
   


### Generating Public and Private keys(Asymmetric encryption)



Generating and using Pub and private keys is one of the asymmetric encryption technology. When party A tries to send message to party B there's a chance that party X intercepts the message. Party A and party B has their own public and private keys. B sends his public key to party A and party A encrypts data using that key and sends the ciphertext. Then party B decrypts the message by using his private key. Party X can't open up the file by his own key.

But this technology is still unsecured. Okay, while party B sends his pub key to A there's a possibility that party X replaced the B's pub key to his own key. Then A don't know which key is B's key. And encrypts data with X's key then sends.Then X decrypts data. So the Message authentication code appears. And the other problem is it takes time.


### Message Authentication code

1. Party A creates a key(by hash function) that will be used to create a MAC
2. Sends the key to party B using secure method
3. Party A creates values using the same key
4. Compare 2 hash of MAC

### Hash function



Hash is function that creates a hash value of any file. All files in computer saved by 1 and 0s. And converting to hash value by mathematical forums that can't go back to original data like mathematic mixer. So the use is comparing hashes. It will be not effected to almost same datas. If only one letter is different from original it will be completely different hash values. The most common hash algorithms are MD5 SHA(secure hash algorithm).
